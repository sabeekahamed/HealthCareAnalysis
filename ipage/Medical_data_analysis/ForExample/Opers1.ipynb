```python
df1=pd.read_csv("Combined_csv.csv",encoding = 'ISO-8859-1')
df1
```

```python
#MANUFACTURER ------------COUNT--------------PRICE======DF2===
lt1=[]
lt2=[]
lt3=[]
i=0
p=0
for sn in df1.Manufacturer:
    if sn not in lt1:
        t=0
        for sk in df1.Manufacturer:#and sp.any() in df1['Price(In Rs.)']:
            if(sn==sk):
                t=t+1
        for sp in df1['Price(In Rs.)'][df1.Manufacturer==sn]:
            p=p+sp
            
        i=i+t
        lt1.append(sn)
        lt2.append(t)
        lt3.append(p)
        p=0
df2=pd.DataFrame({'Manufacturer':lt1,'counts':lt2,'price':lt3})
```

```python
k1=df2[df2.counts==df2.counts.max()]
print("The Manufacturer of highly recommended medicine is:"+k1.Manufacturer.to_string(index=False))
print("\tThe count is:"+k1.counts.to_string(index=False))
print("\tThe price is:"+k1.price.to_string(index=False))
k1=df2[df2.price==df2.price.max()]
print("The Manufacturer of costly recommended medicine is:"+k1.Manufacturer.to_string(index=False))
print("\tThe count is:"+k1.counts.to_string(index=False))
print("\tThe price is:"+k1.price.to_string(index=False))
```
